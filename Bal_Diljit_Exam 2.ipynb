```python
# Setup
import seaborn as sns     
import matplotlib.pyplot as plt
```

```python
import bs4   
print(sns.__version__)

print(sns.get_dataset_names())
```

```python
titanic=sns.load_dataset('titanic')
titanic.info
```

```python
titanic.describe()
```

```python
#Overall distribution of age on Titanic
sns.distplot(a=titanic.age, bins=20)
plt.title("Distribution of Titanic Passenger Age")
```

```python
#Distribution of Class and Age
sns.catplot(x='class', y='age', s=4.2, data=titanic,kind='swarm',order=['First','Second','Third'])
plt.title("Distribution of Titanic Passengers based on Class and Age")
```

```python
#Distribution of Class and Age
sns.catplot(x='sex', y='age', s=4.2, data=titanic,kind='swarm',)
plt.title("Distribution of Titanic Passengers based on Gender and Age")
```

```python
#Survivability based on Age and Gender
gr1=sns.catplot(x='survived', y='age',data=titanic,hue="sex",kind="swarm")
plt.title("Survivability based on Age")
gr1.set_xticklabels(["No","Yes"])
```

```python
#Division of class based on gender and their respective count
sns.countplot(x='class',data=titanic,hue="sex",order=['First','Second','Third'])
plt.title("Division of Class based on Gender and their Respective Count")
```

```python
#
sns.catplot(data=titanic, x="class", y="survived", hue="sex", kind="bar",order=['First','Second','Third'])
plt.title("Division of Class based on Survival and their Respective Gender")

```

```python
gr2 = sns.catplot(data=titanic, x="survived", y="age", hue="sex", kind="bar",)
plt.title("Age based on Survival and their Respective Gender")

gr2.set_xticklabels(["No","Yes"])
```

```python
variable=titanic[['survived','sex', 'age','class']]
sns.pairplot(variable,hue='sex')
```
